# 1. First Model

## 1.1 Generate some data for demonstration purposes

```python
rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
```

code above is generating a line scatter plot

```python
plt.plot(x,y,'o')
```

```python
plt.scatter(x,y)
```

## 1.2 Create our first sklearn model (Linear Regression)

```python
# 1. chose class of model
from sklearn.linear_model import LinearRegression
```

```python
# 2. model hyper parameters: instance
# hyperparameters: parameters that the model will use
model = LinearRegression(fit_intercept=False)
model
```
